### In classification problems, it is a good idea to see the relationship between the input variables and the classifier. In here the input variables are divided into 0-25, 25-50, 50-75 and 75-100 percentile groups. The ratio of classifier in each group is plotted here.  ideally if there is a signal the ratio will change.

```python
#three variables cannot be separated into four groups so they are removed 
fil_data=immune.copy()
pd.set_option('display.max_columns', 500)
haha=fil_data.describe()
haha2=haha.drop(columns=['original_shape_Flatness', 'original_shape_LeastAxisLength','original_glszm_GrayLevelNonUniformity', 'original_glszm_SizeZoneNonUniformity' ])
```

```python
for column, ColData in haha2.iteritems():
    if column=='label':
        continue
    bins=[haha[column].iloc[[3][-1]],haha2[column].iloc[[4][-1]], 
      haha2[column].iloc[[5][-1]], haha2[column].iloc[[6][-1]], haha2[column].iloc[[7][-1]]]
    bins
    group=['0-25 %', '25-50%', '50-75%', '75-100%']
    fil_data[column+'_bin']=pd.cut(fil_data[column], bins, labels=group)
    sorting=pd.crosstab(fil_data[column+'_bin'], fil_data['label'])
    sorting.div(sorting.sum(1).astype(float), axis=0).plot(kind='bar', stacked=True)
    plt.xlabel(column)
    plt.ylabel('percentage')
    
```

####  In this one, we are using the backward elimination to obtain the best possible features. The performance metric used here to evaluate feature performance is pvalue. If the pvalue is above 0.05 then we remove the feature, else we keep it.

```python
tempo=immune.copy()
X=tempo.drop(columns=['label'])
y=immune['label']
```

```python
cols = list(X.columns)
pmax = 1
while (len(cols)>0):
    p= []
    X_1 = X[cols]
    X_1 = sm.add_constant(X_1)
    model = sm.OLS(y,X_1).fit()
    p = pd.Series(model.pvalues.values[1:],index = cols)      
    pmax = max(p)
    feature_with_p_max = p.idxmax()
    if(pmax>0.05):
        cols.remove(feature_with_p_max)
    else:
        break
selected_features_BE = cols
print(selected_features_BE) # printing the features with 
```

```python
#the result indicate that only four features have "significant"effect
x=X[['original_shape_MinorAxisLength', 'original_shape_SurfaceVolumeRatio', 'original_firstorder_10Percentile', 'original_firstorder_Mean']]
```

```python
# A Baysean model wrapped in the Kfold
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.naive_bayes import GaussianNB

i=1 
kf=StratifiedKFold(n_splits=10, random_state=42,shuffle=True)
for train_index, test_index in kf.split(x,y):
    print('\n {} of kfold {}'.format(i,kf.n_splits)) 
    x_train, x_test=x.loc[train_index], x.loc[test_index]
    y_train, y_test=y[train_index], y[test_index]
    model=GaussianNB()
    model.fit(x_train, y_train)
    pred_test=model.predict(x_test)
    score=accuracy_score(y_test, pred_test)
    confmat=confusion_matrix(y_test, pred_test)
    tn, fp, fn, tp = confusion_matrix(y_test, pred_test).ravel()
    spec=tn/(tn+fp)
    sens=tp/(tp+fn)
    print(pred_test)
    print ('accuracy score is %.2f and the  confusion matrix is %s,  sensitivity is %.2f specificity is %.2f' 
           %( score, confmat,  sens, spec))
    from sklearn import metrics
    fpr, tpr, _ = metrics.roc_curve(y_test,  pred_test) 
    auc = metrics.roc_auc_score(y_test, pred_test) 
    plt.figure(figsize=(12,8)) 
    plt.plot(fpr,tpr,label="validation, auc="+str(auc)) 
    plt.xlabel('False Positive Rate') 
    plt.ylabel('True Positive Rate') 
    plt.legend(loc=4) 
    plt.show()
    i=i+1
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
from sklearn.metrics import roc_auc_score
i=1 
kf=StratifiedKFold(n_splits=10, random_state=42,shuffle=True)
for train_index, test_index in kf.split(x,y):
    print('\n {} of kfold {}'.format(i,kf.n_splits)) 
    x_train, x_test=x.loc[train_index], x.loc[test_index]
    y_train, y_test=y[train_index], y[test_index]
    weights={0: 1, 1: 1}
    model=RandomForestClassifier(random_state=1, class_weight=weights, n_estimators=400, max_depth=20)
    model.fit(x_train, y_train)
    pred_test=model.predict(x_test)
    score=accuracy_score(y_test, pred_test)
    confmat=confusion_matrix(y_test, pred_test)
    fscore=f1_score(y_test, pred_test)
    recall=recall_score(y_test, pred_test)
    roc_score=roc_auc_score(y_test, pred_test)
    print ('accuracy score is %.2f and the  confusion matrix is %s, F1 score is %.2f, recall score is %.2f, Roc score is %.2f'
           %( score, confmat, fscore, recall, roc_score))
    from sklearn import metrics
    fpr, tpr, _ = metrics.roc_curve(y_test,  pred_test) 
    auc = metrics.roc_auc_score(y_test, pred_test) 
    plt.figure(figsize=(12,8)) 
    plt.plot(fpr,tpr,label="validation, auc="+str(auc)) 
    plt.xlabel('False Positive Rate') 
    plt.ylabel('True Positive Rate') 
    plt.legend(loc=4) 
    plt.show()
    i=i+1
   
```
